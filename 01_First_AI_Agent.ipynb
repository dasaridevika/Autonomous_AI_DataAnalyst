# Import libraries

```python
from datetime import datetime
```

We import Python's built-in datetime library.
This will help our AI agent know the current date and time.

# Create Tool 1(Caluculator)

```python
def calculator(expression):
    """Performs mathematical calculations."""
    try:
        result=eval(expression)
        return f"Answer : {result}"
    except Exception:
        return "Invalid mathematical expression"
    
```

```python
expression=input("Enter your expression : ")
calculator(expression)
```

# Create Tool 2(Current Time)

```python
def get_time():
    """Returns current data and time."""
    current=datetime.now()
    #gets 2026-06-25 20:15:30
    return current.strftime("%d-%m-%Y %I:%M %p")
    #Format it nicely 25-06-2026 08:15 PM
    
```

```python
get_time()
```

# Create Tool 3 (Knowledge Tool)


```python
knowledge_base={
    "ceo of google": "Sundar Pichai",
    "capital of india": "New Delhi",
    "founder of microsoft": "Bill Gates"
}
def knowledge_tool(question):
    question=question.lower()
    for key,value in knowledge_base.items():
        if key in question:
            return value
    return "Sorry, I don't know that"
```

# Create the Agent Brain

```python
def ai_agent(user_input):
    user_input=user_input.lower()
    # Calculator Tool
    if any(op in user_input for op in ["+","-","*","/","%"]):
        return calculator(user_input)
    #Time Tool
    elif "time" in user_input:
        return get_time()
    # Knowledge Tool
    else:
        return knowledge_tool(user_input)
```

# Run the Agent

```python
while True:
    user=input("You:")
    if user.lower()=="exit":
        print("Agent: Goodbye")
        break
    response =ai_agent(user)
    print("Agent: ",response)
```
