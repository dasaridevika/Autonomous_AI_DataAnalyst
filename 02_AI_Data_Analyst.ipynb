# Import Libraries

```python
import pandas as pd
```

# Load Dataset

```python
df=pd.read_csv("employees.csv")
df.head()
```

# Create Tools

Tool1 --> Show First Rows

```python
def show_head():
    return df.head()
```

```python
show_head()
```

Tool2 --> Dataset Shape

```python
def dataset_shape():
    rows,cols=df.shape

    return f"Rows: {rows}, Columns: {cols}"
```

```python
dataset_shape()
```

Tool3-->Missing Values

```python
def missing_values():
    return df.isnull().sum()
```

```python
missing_values()
```

Tool4 --> Summary Statistics


```python
def summary_stats():
    return df.describe()
```

```python
summary_stats()
```

Too15 --> Average Salary

```python
def average_salary():
    return f"Average Salary: {df['Salary'].mean()}"
```

```python
average_salary()
```

# Build the Agent Brain

```python
def data_agent(question):
    question=question.lower()
    # Show head
    if "head" in question or "first rows" in question:
        return show_head()
    # Shape
    elif "shape" in question or "rows" in question or "columns" in question:
        return dataset_shape()
    #Missing Values
    elif "missing" in question or "null" in question:
        return missing_values()
    # Statistics
    elif "summary" in question or "statistics" in question:
        return summary_stats()
    # Average Salary
    elif "average salary" in question:
        return average_salary()
    
    else:
        return "I cannot understand your question."
         
```

# Run the Agent

```python
while True:
    user_question=input("Ask me about the dataset: ")
    if user_question.lower()=="exit":
        print("Goodbye!")
        break
    answer=data_agent(user_question)
    
    print(answer)
```
